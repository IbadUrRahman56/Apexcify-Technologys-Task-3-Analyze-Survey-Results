### TASK 3: Analyze Survey Results

### Cell 1 — Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Cell 3 — Load CSV File

```python
df = pd.read_csv("survey_results.csv")
df.head
```

### Cell 3 — Count Satisfaction Levels

```python
counts = df["Satisfaction"].value_counts()
print("Satisfaction Level Counts:")
print(counts)
```

### Cell 4 — Plot Bar Chart

```python
 plt.figure(figsize=(8, 5))
sns.barplot(x=counts.index, y=counts.values, palette="viridis")

plt.xlabel("Satisfaction Level", fontsize=12)
plt.ylabel("Number of People", fontsize=12)
plt.title("Survey Results - Satisfaction Levels", fontsize=14)
plt.xticks(rotation=30)
plt.show()
```
